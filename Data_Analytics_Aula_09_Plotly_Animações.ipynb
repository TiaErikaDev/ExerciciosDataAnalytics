```python
import pandas as pd
```

```python
projecao = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/projecao_IBGE_2018/projecao_IBGE_2018_atualizada06042020.csv',
    sep = ';',
    decimal = ',',
    encoding = 'utf-8'
)
projecao
```

```python
projecao.info(verbose=True)
```

```python
#Filtrando apenas dados das unidades da federação
projecao_uf = projecao.query('Codigo_UF!=0')
projecao_uf
```

```python
import plotly.express as px
```

```python
px.scatter(
    data_frame = projecao_uf,
    x = 'TMI',
    y = 'EVN',
    animation_frame = 'Ano',
    animation_group = 'Nome_UF',
    size = 'Populacao_Total',
    color = 'Nome_Regiao',
    hover_name = 'Nome_UF',
    #Tamanho das bolinhas
    size_max = 50,
    range_x = [0,40],
    range_y = [60,90]
)
```

##Gráfico de barras animado

```python
projecao_regioes = projecao.query('(Codigo_UF==0) & (Codigo_Regiao!=0)')
projecao_regioes
```

```python
projecao_regioes.info(verbose=True)
```

```python
fig = px.bar(
    data_frame = projecao_regioes,
    x = 'Nome_Regiao',
    y = 'TMI',
    color = 'Nome_Regiao',
    animation_frame = 'Ano',
    animation_group = 'Nome_Regiao',
    title = 'Taxa de mortalidade infantil por região do Brasil (2010-2060)',
    labels={
        'Nome_Regiao':'Grande região do Brasil',
        'TMI':'Taxa de Mortalidade Infantil'
    },
    range_y = [0,30]
)
```

```python
fig.show()
```

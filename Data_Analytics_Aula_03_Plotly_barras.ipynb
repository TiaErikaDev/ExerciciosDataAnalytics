```python
#importando o pandas
import pandas as pd
```

```python
#importando o conjunto de dados BNDES
bndes = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/bndes/operacoes-financiamento-operacoes-nao-automaticas.csv',
    sep = ';',
    decimal = ',',
    encoding = 'utf-8'
)

bndes
```

```python
#informações do conjunto de dados
bndes.info(verbose=True)
```

```python
#verificando valores missing (perdidos)
bndes.isnull().sum()
```

```python
#Transformando data_da_contratação em formato de data e colocando em uma variavel nova (data)
bndes['data'] = pd.to_datetime(bndes.data_da_contratacao, infer_datetime_format=True)
```

```python
#Extraindo o ano da variavel data
bndes['ano'] = pd.DatetimeIndex(bndes['data']).year
```

```python
#Verificando a nova estrutura do dataset
bndes.info()
```

```python
#Observando o conjunto de dados
bndes.head(2)
```

```python
#Filtrando somente os financiamentos que ocorreram a partir de 2018
bndes_2018 = bndes.query('ano==2018')
```

```python
#Espiando oconjunto de dados
bndes_2018.ano.value_counts()
```

```python
#importanto as bibliotecas 
import plotly.express as px
```

###Quantidade de financiamentos por setor do BNDES

> Bloco com recuo



```python
from pandas._libs.hashtable import value_count
#Contando as operações por setor e colocando em uma nova tabela (DataFrame-df)
df_bndes_setor = bndes_2018[['setor_bndes']]\
                      .value_counts()\
                      .to_frame()\
                      .reset_index()\
                      .rename(columns={0: 'financiamentos'})
#Espiando o resultado
df_bndes_setor
```

```python
#Construindo um gráfico de colunas da quantidade de financiamentos por setor
px.bar(
    data_frame = df_bndes_setor,
    x = 'setor_bndes',
    y = 'financiamentos',
)
```

###Quantidade de financiamentos por setor do BNDES e inovação

> Bloco com recuo




```python
#contando as operações por setor e inovação
df_setor_inovacao = bndes_2018[['setor_bndes', 'inovacao']]\
                        .value_counts()\
                        .to_frame()\
                        .reset_index()\
                      .rename(columns={0:'financiamentos'})

#Espiando o resultado
df_setor_inovacao
```

```python
#Construindo um gráfico de colunas da quantidade de financiamentos por setor e inovação
px.bar(
    data_frame = df_setor_inovacao,
    x='setor_bndes',
    y='financiamentos',
    color='inovacao'
)
```

```python
#fazendo algumas mudanças no gráfico anterior
px.bar(
    data_frame = df_setor_inovacao,
    x='setor_bndes',
    y='financiamentos',
    color='inovacao',
    barmode='group',
    text_auto=True
)
#PS: O barmode do gráfico anterior era 'relative
```

```python
#fazendo algumas mudanças no gráfico anterior
px.bar(
    data_frame = df_setor_inovacao,
    x='setor_bndes',
    y='financiamentos',
    color='inovacao',
    barmode='overlay',
)
#nesse caso a visualização fica sobreposta
```

```python
#fazendo algumas mudanças no gráfico anterior
px.histogram(
    data_frame = df_setor_inovacao,
    x='setor_bndes',
    y='financiamentos',
    color='inovacao',
    barmode='relative',
    barnorm='percent',
    color_discrete_map= {
        'NÃO' : 'lightblue',
        'SIM' : 'purple',
    },
    title = 'Percentual do financiamento do BNDES para inovação, em 2018, por setor'
)
```

###Valor desembolsado nos financiamentos por setor do BNDES e inovação


```python
#Gerando df do valor desembolsado por UF
df_bndes_uf_valor = bndes_2018[['uf', 'valor_desembolsado_reais']]\
                      .groupby('uf', as_index=False)\
                      .sum()

#Visualizando o resultado
df_bndes_uf_valor
```

```python
#Construindo o gráfico de barras
px.bar(
    data_frame = df_bndes_uf_valor,
    y = 'uf',
    x = 'valor_desembolsado_reais',
    orientation = 'h'
)
```

```python
#Construindo o gráfico de barras com ajustes
grafico_barras_valor_uf = px.bar(
    data_frame = df_bndes_uf_valor,
    y = 'uf',
    x = 'valor_desembolsado_reais',
    orientation = 'h',
    title = 'Total desembolsado em reais nos financiamentos do BNDES, em 2018, por UF.',
    height = 800,
    width = 700,
    labels = {
        'uf' : 'Unidade da Federação',
        'valor_desembolsado_reais' : 'Valor desembolsado em reais'
    }
)

#Organizando em ordem decrescente
grafico_barras_valor_uf.update_layout(yaxis={'categoryorder' : 'total ascending'})

#Alterando as cores do gráfico
grafico_barras_valor_uf.update_traces(
    marker_color = 'green',
    marker_line_color = 'black',
    marker_line_width = 0.5,
    opacity=0.9
)

#Exibindo o gráfico
grafico_barras_valor_uf
```

```python
#importando o pandas
import pandas as pd
```

```python
#Importando o conjunto de dados do ideb 2019 
ideb = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/INEP/ideb_ensino_medio_escolas_2019.csv', 
    sep=';',
    decimal=',',
    encoding='utf-8',
    na_values='-'
)
```

```python
#Visualizando os dados
ideb
```

```python
ideb.info(verbose=True)
```

```python
#Verificando se tem dados faltando
ideb.isnull().sum()
```

##Transformações no conjunto de dados

```python
#Retirando os dados ausentes
ideb_medio = ideb[
    (ideb["nota_SAEB_2019_matematica"].notna()) & \
    (ideb["nota_SAEB_2019_lingua_portuguesa"].notna()) & \
    (ideb["rede"].notna())
]
```

```python
#Verificando valores missing
ideb_medio.isnull().sum()
```

##Histograma das notas de Português e Matemática em 2019

```python
#importando as bibliotecas necessárias
import plotly.express as px
```

```python
#Construindo histogramas
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica'
)
```

```python
#limitando o número de bins
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    nbins = 20
)
```

```python
#Mudando escala de contagem para probabilidade
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    histnorm = 'probability'
)
```

```python
#Estratificando por rede de ensino
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    histnorm = 'probability',
    color = 'rede'
)
```

```python
#Inserindo gráficos marginais com as distribuições dos grupos de estratificação
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    nbins = 50,
    histnorm = 'probability',
    color = 'rede',
    marginal = 'box'
)
```

```python
#separando a visualização em colunas
px.histogram(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    nbins = 50,
    histnorm = 'probability',
    color = 'rede',
    marginal = 'box',
    facet_col = 'rede'
)
```

##ECDF - Empirical cumulative distribution function da nota de Português e Matemática em 2019


```python
#Construindo a ECDF da nota de Matemática
px.ecdf(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica'
)
#Mostra a probabilidade acumulada de ocorrer uma nota no intervalo X
```

```python
#Construindo a ECDF da nota de Matemática
px.ecdf(
    data_frame = ideb_medio,
    x = 'nota_SAEB_2019_matematica',
    color = 'rede'
)
```

```python
#Construindo a ECDF da nota de Português e Matemática
px.ecdf(
    data_frame = ideb_medio,
    x = ['nota_SAEB_2019_matematica', 'nota_SAEB_2019_lingua_portuguesa']
)
```

```python
#Estratificando por rede de ensino
px.ecdf(
    data_frame = ideb_medio,
    x = ['nota_SAEB_2019_matematica', 'nota_SAEB_2019_lingua_portuguesa'],
    facet_col = 'rede'
)
```

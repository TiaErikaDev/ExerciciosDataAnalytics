```python
import pandas as pd
```

```python
criptomoedas = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/criptomoedas/criptomoedas.csv',
    sep=';',
    decimal='.'
)

criptomoedas
```

```python
criptomoedas.info(verbose = True)
```

```python
criptomoedas.isnull().sum()
```

```python
criptomoedas['Data'] = pd.to_datetime(criptomoedas.Data, infer_datetime_format=True)
```

```python
import plotly.express as px
```

```python
#pegando somente a série do Bitcoin
bitcoin = criptomoedas.query('Iniciais=="BTC-USD"')
```

```python
#Gerando o gráfico de linhas com os maiores preços de Bitcoin por dia
px.line(
    data_frame=bitcoin,
    x='Data',
    y='Alta'
)
```

```python
#incrementando o gráfico de linhas anterior
px.line(
    data_frame=bitcoin,
    x='Data',
    y='Alta',
    title='<b>Maior preço de Bitcoin por dia.</b>',
    labels = {
        'Data' : 'Dia da Tansação',
        'Alta' : 'Maior valor do dia (em dolar)'
    },
    markers=True,
    template='ggplot2'
)
```

```python
#Mudando a cor da linha do gráfico anterior
px.line(
    data_frame=bitcoin,
    x='Data',
    y='Alta',
    title='<b>Maior preço de Bitcoin por dia.</b>',
    labels = {
        'Data' : 'Dia da Tansação',
        'Alta' : 'Maior valor do dia (em dolar)'
    },
    color='Iniciais',
    color_discrete_map={
        'BTC-USD' : 'yellow'
    },
    template='plotly_dark'
)
```

```python
#Preços mínimos
px.line(
    data_frame = criptomoedas,
    x='Data',
    y='Baixa',
    color='Iniciais',
    template='plotly_white'
)
```

```python
#Incrementando o gráfico anterior
px.line(
    data_frame = criptomoedas,
    x='Data',
    y='Baixa',
    color='Iniciais',
    color_discrete_map={
        'BTC-USD' : 'green',
        'ETH-USD' : 'orange'
    },
    title = '<b>Preços mínimos por dia de Criptomoedas.</b>',
    labels= {
        'Data' : 'Dia da Transação',
        'Alta' : 'Menor valor do dia (em Dolar)'
    },
    markers=True,
    symbol='Iniciais',
    template='presentation'
)
```

```python
#Ajustando a escala para logarítimica
px.line(
    data_frame = criptomoedas,
    x='Data',
    y='Baixa',
    color='Iniciais',
    color_discrete_map={
        'BTC-USD' : 'darkblue',
        'ETH-USD' : 'darkorange'
    },
    title = '<b>Preços mínimos por dia de Criptomoedas.</b>',
    labels= {
        'Data' : 'Dia da Transação',
        'Baixa' : 'Menor valor do dia (em Dolar)'
    },
    template='simple_white',
    log_y = True
)
```

```python
petro = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/mercado_acoes/PETR4.csv',
    sep=',',
    decimal='.'
)

petro
```

```python
petro.info(verbose=True)
```

```python
petro['Date'] = pd.to_datetime(petro.Date, infer_datetime_format=True)
```

```python
px.line(
    data_frame=petro,
    x='Date',
    y='High',
    template='plotly_white'
)
```

```python
px.line(
    data_frame=petro,
    x='Date',
    y='Low',
    template='plotly_white'
)
```

```python
bolsonaro = petro.query('Date>="2019"')
bolsonaro
```

```python
lula = petro.query('"2003"<=Date<="2011"')
lula
```

```python
px.line(
    data_frame=lula,
    x='Date',
    y='High',
    template='plotly_white'
)
```

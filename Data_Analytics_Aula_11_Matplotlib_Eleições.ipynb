```python
import pandas as pd
```

```python
resultados = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/eleicoes_2018/resultados_candidato_2018.csv',
    sep = ';',
    decimal = ',',
    encoding = 'latin-1'
)
```

```python
resultados.head(5)
```

```python
candidatos = pd.read_csv(
    filepath_or_buffer = 'candidatos_2018.csv',
    sep = ';',
    decimal = ',',
    encoding = 'latin-1'
)
```

```python
candidatos.head(5)
```

```python
resultados.info(verbose=True)
```

```python
candidatos.info(verbose=True)
```

```python
#filtrando apenas algumas informações
candidatos_para_merge = candidatos[['sequencial', 'nome', 'nome_urna', 'numero_partido',\
                                    'situacao', 'ocupacao', 'genero', 'idade', 'instrucao', 'estado_civil',\
                                    'nacionalidade', 'sigla_uf_nascimento', 'municipio_nascimento',\
                                    'raca']]
```

```python
#Juntando as bases de dados
eleicoes = pd.merge(
    left = resultados,
    right = candidatos_para_merge,
    how = 'left',
    left_on = 'sequencial_candidato',
    right_on = 'sequencial'
)
```

```python
eleicoes.info(verbose=True)
```

```python
eleicoes
```

```python
#importando as bibliotecas necessárias
import matplotlib.pyplot as plt
```

```python
#Distribuição por sexo
perfil_sexo = eleicoes.genero.value_counts()\
    .rename_axis('sexo') \
    .to_frame('candidatos') \
    .reset_index()

perfil_sexo
```

```python
plt.pie(x=perfil_sexo.candidatos, labels=perfil_sexo.sexo);
```

```python
#Inserir as porcentagens
plt.pie(x=perfil_sexo.candidatos, labels=perfil_sexo.sexo, autopct = '%.2f%%');
```

```python
#Customizando o gráfico
plt.pie(
    x=perfil_sexo.candidatos, 
    labels=perfil_sexo.sexo, 
    autopct = '%.2f%%',
    shadow = True,
    colors = ['blue', 'magenta'],
    explode = [0,0.1],
    startangle = 90,
    textprops = dict(color='white')
    );
```

```python
#Definindo uma função para customizar os rótulos
def autopct_custom(valor):
    def autopct_def(pct):
        total = sum(valor)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% (N=(v:d))'.format(p=pct,v=val)
    return autopct_def

#Definindo o tamanho da figura
plt.figure(figsize=[10,7])

#Fazendo o gráfico de pizza
plt.pie(
    x=perfil_sexo.candidatos,
    labels=perfil_sexo.sexo,
    autopct=autopct_custom(perfil_sexo.candidatos),
    shadow=True,
    colors=['lightgray','orange'],
    explode=[0,0.1],
    startangle=90,
    textprops=dict(color='black')
)

#inserindo o título
plt.title('Sexo dos candidatosdas eleições 2018');
```

```python
#customizando o rótulo dos valores com a função lambda

#Definindo o tamanho da figura
plt.figure(figsize=[10,7])

#Fazendo o gráfico de pizza
plt.pie(
    x=perfil_sexo.candidatos,
    labels=perfil_sexo.sexo,
    autopct=lambda p : '{:.2f}% (N={:.0f})'.format(p,p * sum(perfil_sexo.candidatos)/100) ,
    shadow=True,
    colors=['darkgreen','darkorange'],
    explode=[0,0.1],
    startangle=90,
    textprops=dict(color='white')
)

#inserindo o título
plt.title('Sexo dos candidatosdas eleições 2018');
```

##Distribuição dos candidatos por estado civil

```python
#distribuição dos candidatos por cargo
perfil_estcivil = eleicoes.estado_civil.value_counts() \
    .rename_axis('estado_civil')\
    .to_frame('candidatos')\
    .reset_index()

#Cauculando a porcentagem
perfil_estcivil['pct'] = (perfil_estcivil['candidatos'] / perfil_estcivil['candidatos'].sum()) \
                   .mul(100).round(2)

perfil_estcivil
```

```python
#Fazendo um gráfico de barras
plt.barh(perfil_estcivil.estado_civil , perfil_estcivil.candidatos);
```

```python
#Customizando o gráfico de barras
#Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(16,9))

#Fazendo o gráfico de colunas
ax.barh(
    y = perfil_estcivil.estado_civil,
    width = perfil_estcivil.pct,
    height = 0.8,
    color = 'maroon',
    edgecolor = 'black'
);

#inserindo o título
ax.set_title('Estado civil dos candidatos das eleições 2018', loc = 'left')

#Invertendo o eixo Y
ax.invert_yaxis()

#inserindo rótulos dos eixos
plt.xlabel("Proporção de candidados (em %)")
plt.ylabel('Estado Civil')

plt.show()
```

##Distribuição dos candidatos por escolaridade

```python
#distribuição dos candidatos por cargo
perfil_instrucao = eleicoes.instrucao.value_counts() \
    .rename_axis('instrucao')\
    .reindex([
        'le e escreve',
        'ensino fundamental incompleto',
        'ensino fundamental completo',
        'ensino medio incompleto',
        'ensino medio completo',
        'ensino superior incompleto',
        'ensino superior completo']) \
    .to_frame('candidatos') \
    .reset_index()

#Calculando porcentagem
perfil_instrucao['pct'] = (perfil_instrucao['candidatos']/perfil_instrucao['candidatos'].sum()) \
                    .mul(100).round(2)

perfil_instrucao
```

```python
#Customizando o gráfico de barras
#Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(16,9))

#Fazendo o gráfico de colunas
ax.bar(
    x = perfil_instrucao.instrucao,
    height = perfil_instrucao.candidatos,
    width = 0.8,
    color = 'orange',
    edgecolor = 'black'
);

#inserindo o título
ax.set_title('Grau de instrução dos candidatos das eleições 2018', loc = 'left')

#inserindo rótulos dos eixos
plt.xlabel("Grau de instrução")
plt.ylabel('Número de candidatos')

#Adicionando rótulo de valores
for i in ax.patches:
    plt.text(
        x = i.get_x()+0.3,
        y = i.get_height()+50,
        s = str(round((i.get_height()), 2)),
        fontsize = 10,
        fontweight = 'bold',
        color = 'black',
    )

#Rotacionando os rótulos do eixo x
plt.xticks(rotation=45, ha='right')

plt.show()
```

##Raça/cor dos candidatos

```python
#Distribuição por sexo
perfil_raca = eleicoes.raca.value_counts() \
    .rename_axis('raca') \
    .reindex([
        'amarela',
        'branca',
        'indigena',
        'parda',
        'preta']) \
    .to_frame('candidatos') \
    .reset_index()

#Calculando a poecentagem
perfil_raca['pct'] = (perfil_raca['candidatos'] / perfil_raca['candidatos'].sum()) \
                .mul(100).round(2)

perfil_raca
```

```python
#Customizando o gráfico de colunas
#Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(16,9))

#Fazendo o gráfico de colunas
ax.bar(
    x = perfil_raca.raca,
    height = perfil_raca.pct,
    width = 0.8,
    color = 'purple',
    edgecolor = 'black'
);

#inserindo o título
ax.set_title('Raça/Cor dos candidatos das eleições 2018', loc = 'left')

#inserindo rótulos dos eixos
plt.xlabel("Raça/Cor")
plt.ylabel('Proporção dos candidatos (em %)')

#Adicionando rótulo de valores
for i in ax.patches:
    plt.text(
        x = i.get_x()+0.3,
        y = i.get_height()+1,
        s = str(round((i.get_height()), 2)),
        fontsize = 10,
        fontweight = 'bold',
        color = 'black',
    )

plt.show()
```

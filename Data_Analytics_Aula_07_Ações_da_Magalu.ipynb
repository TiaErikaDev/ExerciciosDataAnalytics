```python
import pandas as pd
```

```python
magalu = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/mercado_acoes/MGLU3.csv',
    sep = ',',
    decimal = '.'
)
magalu
```

```python
magalu.info(verbose=True)
```

```python
magalu.isnull().sum()
```

```python
magalu['Date'] = pd.to_datetime(magalu.Date, infer_datetime_format=True)
```

```python
#Colocando a coluna Date como índex do conjunto de dados
magalu.set_index(keys='Date', inplace=True)
```

```python
magalu.info(verbose=True)
```

```python
magalu
```

```python
import plotly.graph_objects as go
```

```python
grafico_estilo_bovespa = go.Figure(
    data=[
        go.Candlestick(
            x=magalu.index,
            open=magalu['Open'],
            high=magalu['High'],
            low=magalu['Low'],
            close=magalu['Close']
        )
    ]
)
```

```python
grafico_estilo_bovespa.update_layout(xaxis_rangeslider_visible = False)
```

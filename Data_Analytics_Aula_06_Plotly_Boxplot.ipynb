```python
import pandas as pd
```

```python
rh = pd.read_csv(
    filepath_or_buffer = 'https://raw.githubusercontent.com/jonates/opendata/master/turnover/turnover_treino.csv',
    sep = ';',
    decimal = ','
)
```

```python
rh
```

```python
rh.info(verbose = True)
```

```python
rh.isnull().sum()
```

```python
#Analisando os percentis
rh['contrato_salario'].describe(percentiles=[.001, .01, .05, .1, .25, .5, .75, 0.9, 0.95, 0.99])
```

##Gráfico dos salários

```python
#importando as bibliotecas necessárias
import plotly.express as px
```

```python
rh_eda = rh.query('contrato_salario<4593.520500')
```

```python
#Boxplot dos salários de 95% do conjunto de dados
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    width = 800,
    height = 600
)
```

```python
#Estratificando a analise dos salários pelo sexo dos trabahadores
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    x = 'func_sexo',
    template = 'ggplot2',
    width = 800,
    height = 600
)
```

```python
#Estratificando a analise dos salários pela raça/cor dos trabahadores
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    color = 'func_racacor',
    title = 'Distribuição dos salários por raça/cor do trabalhador',
    template = 'plotly_white'
)
```

```python
#Estratificando a analise dos salários por existência de deficiência dos trabahadores
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    color = 'func_deficiencia',
    color_discrete_map = {
        'NÃO' : 'red',
        'SIM' : 'green'
    },
    title = 'Distribuição dos salários por existência de deficiência dos trabahadores',
    labels = {
        'contrato_salario' : 'Salário (em reais)',
        'func_deficiencia' : 'Trabalhador possui alguma deficiência'
    },
    template = 'none',
    width = 800,
    height = 600
)
```

```python
#Estratificando a analise dos salários pela escolaridade dos trabahadores
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    color = 'func_escolaridade',
    title = 'Distribuição dos salários por scolaridade dos trabahadores',
    labels = {
        'contrato_salario' : 'Salário (em reais)',
        'func_escolaridade' : 'Escolaridade'
    },
    template = 'plotly_dark',
    category_orders = {'func_escolaridade':['ANALFABETO-FUND_INCOMPLETO', 'FUND_COMPLETO-MEDIO_INCOMPLETO', 'MEDIO COMPLETO', 'SUPERIOR_INCOMPLETO', 'SUPERIOR_COMPLETO']}
)
```

```python
#Estratificando a analise dos salários por UF dos trabahadores
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    color = 'func_uf',
    title = 'Distribuição dos salários por UF dos trabahadores',
    labels = {
        'contrato_salario' : 'Salário (em reais)',
        'func_escolaridade' : 'UF'
    },
    template = 'presentation',
)
```

```python
#Estratificando a analise dos salários por setor da economia das empresas
px.box(
    data_frame = rh_eda,
    y = 'contrato_salario',
    color = 'empresa_setor',
    title = 'Distribuição dos salários por setor das empresas',
    labels = {
        'contrato_salario' : 'Salário (em reais)',
        'func_escolaridade' : 'setor'
    },
    template = 'xgridoff',
)
```

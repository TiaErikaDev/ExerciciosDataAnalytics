```python
#Importando o conjunto de dados da Netflix e colocando dentro do objeto chamado netflix
netflix = pd.read_csv(
    filepath_or_buffer = '/content/netflix_titles.csv', 
    sep=',',
    encoding = 'utf-8'
)
```

Gráfico em pizza

```python
#importando as bibliotecas necessárias
import plotly.express as px
```

```python
#construindo o gráfico de pizza com o tipo das produções
px.pie(data_frame = netflix, names = 'type')
```

```python
#verificando se o plotly fez a distribuição correta
netflix['type'].value_counts(normalize=True).mul(100).round(1)
```

```python
#customizando as cores do gráfico de pizza através de escalas prontas
px.pie(
    data_frame = netflix,
    names = 'type',
    color = 'type',
    color_discrete_sequence=px.colors.sequential.Aggrnyl_r  
)
```

```python
#customizando as cores do gráfico de pizza através de escalas customizadas
px.pie(
    data_frame = netflix,
    names = 'type',
    color = 'type',
    color_discrete_map = {
        'movie' : 'gray',
        'TV Show' : 'orange',
    } 
)
```

```python
#Outras customizações de um gráfico de pizza
px.pie(
    data_frame = netflix,
    names = 'type',
    color = 'type',
    color_discrete_map = {
        'movie' : 'green',
        'TV Show' : 'pink',
    },
    title = 'Distribuição da programação da netflix por tipo',
    labels = {'type':'Tipo da Produção'},
    opacity = 1,
    width=600,
    height=400
)
```

```python
#construindo o gráfico de rosca
px.pie(
    data_frame = netflix,
    names = 'type',
    color = 'type',
    color_discrete_map = {
        'movie' : 'gray',
        'TV Show' : 'red',
    },
    title = '<b>Netflix</b><br>Distribuição da programação da netflix por tipo',
    labels = {'type':'Tipo da Produção'},
    opacity = 1,
    width=800,
    height=600,
    hole = .7
)
```
